# 1.2.2

```python
def stop(a, b, e):
    return (abs(b) - abs(a) < e)

def grad(x):
    return -2*x

def f(x):
    return -1*(x**2)

def gradientAscent(x0, p, e, R):
    curMax = x0
    prevMax = x0
    for step in range(0, R):
        prevMax = curMax
        if step == 0:
            curMax = grad(prevMax)
        else:
            curMax = prevMax + (p * grad(prevMax))
            print('------------')
            print("step: ", step)
            print("x: ", prevMax)
            print("f(x): ", f(curMax))
            print("grad f(x): ", curMax)
            if stop(curMax, prevMax, e):
                break

gradientAscent(10, 0.1, 0.5, 100)
```

# 1.2.3

```python
# Parameters:
#   gradientAscent(x0, p, e, R)
#   where...
#      x0 is initial value
#      p is learning rate
#      e is small number for stopping criteria
#      R is bail-out number for stopping criteria

# Decrease e
gradientAscent(10, 0.1, 1.0, 100) # 8 steps, 3.355
gradientAscent(10, 0.1, 0.5, 100) # 11 steps, -1.717
gradientAscent(10, 0.1, 0.2, 100) # 15 steps, -0.703
gradientAscent(10, 0.1, 0.05, 100) # 21 steps, -0.184
# Result: More steps and we get closer to the maximum

# Increase initial value
gradientAscent(3, 0.1, 0.5, 100) # 5 steps, -1.966
gradientAscent(8, 0.1, 0.5, 100) # 10 steps, -1.718
gradientAscent(12, 0.1, 0.5, 100) # 12 steps, -1.649
# Result: More steps and we get closer to the maximum

# Increase learning rate
gradientAscent(10, 0.1, 0.5, 100) # 11 steps, -1.717
gradientAscent(10, 0.4, 0.5, 100) # 4 steps, -0.0319
gradientAscent(10, 0.8, 0.5, 100) # 7 steps, 0.5598
gradientAscent(10, 2.0, 0.5, 100) # 1 step, 60.0
# Result: Less steps. Can get closer to max, but has chance of overshooting maximum
```

```python
R = 1000 # repetitions
N = 100 # sample size

m = []

for step in range(0, R):
    sample = np.random.standard_normal(size=100)
    mean = np.mean(sample)
    m.append(mean)

mm, sigma = np.mean(m), np.std(m)

interval = st.norm.interval(0.95, loc=mm, scale=sigma)

print(mm)
print(interval)

# histogram
n, bins, patches = plt.hist(m, 50, normed=1, facecolor='green', alpha=0.75)
plt.grid(True)
plt.plot([interval[0], interval[0]], [0, 5], color='k', linestyle='-', linewidth=2)
plt.plot([interval[1], interval[1]], [0, 5], color='k', linestyle='-', linewidth=2)

plt.show()

count = 0

for mean in m:
    if ((mean < interval[0]) or (mean > interval[1])):
        count = count + 1

print("Number of m-s outside of 95% confidence interval: ", count)
```
