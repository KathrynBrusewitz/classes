# 2 Which variable gives the best branch

## 2.1
Pick the first attribute (which happens to be voting for handicapped infants bill, see the readme file). Split your data according to yea or nay vote on that bill. You can just ignore the other here. You get two subsets of data: yea-sayers and nay-sayers.

```python
# Extract party from first column, decisions from second column
bill = data[[0, 1]]
no = bill[bill[1] == 'n']
yes = bill[bill[1] == 'y']
```

## 2.2
Compute the percentage of republicans and democrats in both of your subsets.

```python
dNo = len(no[no[0] == 'democrat'])/len(no)
dYes = len(yes[yes[0] == 'democrat'])/len(yes)
rNo = len(no[no[0] == 'republican'])/len(no)
rYes = len(yes[yes[0] == 'republican'])/len(yes)

print('Democrat nay-sayers: ', np.round(dNo, 2))
print('Democrat yea-sayers: ', np.round(dYes, 2))
print('Republican nay-sayers: ', np.round(rNo, 2))
print('Republican yea-sayers: ', np.round(rYes, 2))
```

## 2.3 Compute entropy for yea and nay-voters

```python
def heat(a, b):
    return -1 * a * math.log2(a) - b * math.log2(b)

print('Entropies for each branch: ')
print('Nay-sayers: ', heat(dNo, rNo))
print('Yea-sayers: ', heat(dYes, rYes))
```

## 2.4
Compute the weighted average entropy for each potential split

```python
def entropy(column):
    bill = data[[0, column]]
    no = bill[bill[column] == 'n']
    yes = bill[bill[column] == 'y']
    
    total = len(no) + len(yes)
    
    dNo = len(no[no[0] == 'democrat'])/len(no)
    dYes = len(yes[yes[0] == 'democrat'])/len(yes)
    rNo = len(no[no[0] == 'republican'])/len(no)
    rYes = len(yes[yes[0] == 'republican'])/len(yes)

    heatYes = heat(dYes, rYes)
    heatNo = heat(dNo, rNo)
    return (heatYes * len(yes) + heatNo * len(no)) / total
```

```python
for i in data.columns:
    if (i != 0):
        print("Column ", i, ": ", np.round(entropy(i), 2))
```

## 2.5
Which feature gives the smallest average entropy?

```python
minIndex = 1
minEntropy = entropy(1)

for i in data.columns:
    if (i != 0):
        if (entropy(i) < minEntropy):
            minEntropy = entropy(i)
            minIndex = i

print("Column ", minIndex, ": ", np.round(minEntropy, 2))
```
